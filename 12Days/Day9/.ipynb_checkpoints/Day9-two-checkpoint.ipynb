```python
def convert_date(val):
    d, m, y = val.split('-')
    m = months[m]
    return datetime(int ('20' + y), m, int(d))

fec.contb_receipt_dt
fec['contb_receipt_dt'] = fec.contb_receipt_dt.map(convert_date)
```
